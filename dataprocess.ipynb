# Plot Stats

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('data/tokenization_types.csv')

labels = df['Type']
counts = df['Count']

# Tableau (tab) colors
tab_colors = [
    'tab:blue',
    'tab:orange',
    'tab:green',
    'tab:gray',
    'tab:purple',
    'tab:brown'
]

plt.figure()
plt.pie(
    counts,
    labels=labels,
    colors=tab_colors[:len(counts)],
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={
        'edgecolor': 'white',
        'linewidth': 1
    }
)

plt.title('Distribution of Tokenization Types')
plt.axis('equal')

plt.tight_layout()
plt.savefig('types.pdf', bbox_inches='tight', pad_inches=0.05)
plt.show()
```
